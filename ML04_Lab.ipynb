```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("./Corr_Pearson.csv")
#print(df)
df=df.iloc[:, 1:]
df.head()
```

```python
#ploting

plt.scatter(df['x'], df['a'])
plt.scatter(df['x'], df['b'])
plt.scatter(df['x'], df['c'])
plt.scatter(df['x'], df['d'])
plt.scatter(df['x'], df['e'])
```

```python
import math
  
# function that returns correlation coefficient.
def correlationCoefficient(X, Y, n) :
    Covarriant = 0
    squareSum_X = 0
    squareSum_Y = 0

    mean_X=np.mean(X)  
    mean_Y=np.mean(Y)  
      
    i = 0
    while i < n :     
        # sum of X[i]-mean_X * Y[i]-mean_Y.
        Covarriant = Covarriant + (X[i]-mean_X) * (Y[i]-mean_Y)
          
        # sum of square of array elements.
        squareSum_X = squareSum_X + (X[i]-mean_X) * (X[i]-mean_X)
        squareSum_Y = squareSum_Y + (Y[i]-mean_Y) * (Y[i]-mean_Y)
          
        i = i + 1
       
    # use formula for calculating correlation 
    # coefficient.
    corr = (float)(Covarriant)/(float)(math.sqrt(squareSum_X*squareSum_Y))
    return corr
```

```python

X = df['x'].to_numpy()
Y = df['a'].to_numpy()
n = len(X)
print("pearson: x, a")
print(correlationCoefficient(X, Y, n))
df.corr(method='pearson')
```
