```python
import requests

img_url = 'https://glider-it.com/images/header_models/jeemo_1.jpg'
response = requests.get(img_url)
if response.status_code:
    fp = open('image.png', 'wb')
    fp.write(response.content)
    fp.close()
```

```python
# importing the modules needed
import cv2
import numpy as np
  
# Reading the image
image = cv2.imread('image.png')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)



# Creating the kernel(2d convolution matrix)
kernel1 = np.ones((5, 5), np.float32)/30
kernel2 = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
```

```python
gray_image.shape
```

Original Image

```python
import matplotlib.pyplot as plt
plt.imshow(gray_image, cmap='gray')
```

```python
# Applying the filter2D() function
img = cv2.filter2D(src=gray_image, ddepth=-1, kernel=kernel2)
plt.imshow(img,cmap='gray')
```
